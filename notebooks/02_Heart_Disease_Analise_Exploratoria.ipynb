## Contexto do Dataset

Este dataset contém dados de pacientes submetidos a exames cardiológicos. É uma compilação de dados de várias instituições médicas e é amplamente utilizado para estudos de doenças cardiovasculares.

**Variáveis:**
- `age`: Idade (anos)
- `sex`: Sexo (1 = masculino, 0 = feminino)
- `cp`: Tipo de dor no peito (0-3)
- `trestbps`: Pressão arterial em repouso (mm Hg)
- `chol`: Colesterol sérico (mg/dl)
- `fbs`: Glicemia em jejum > 120 mg/dl (1 = verdadeiro, 0 = falso)
- `restecg`: Resultados eletrocardiográficos em repouso (0-2)
- `thalach`: Frequência cardíaca máxima alcançada
- `exang`: Angina induzida por exercício (1 = sim, 0 = não)
- `oldpeak`: Depressão ST induzida por exercício
- `slope`: Inclinação do segmento ST de pico do exercício (0-2)
- `ca`: Número de vasos principais coloridos por fluoroscopia (0-3)
- `thal`: Talassemia (1 = normal, 2 = defeito fixo, 3 = defeito reversível)
- `target`: Diagnóstico de doença cardíaca (0 = sem doença, 1-4 = com doença)


## 1. Importação de Bibliotecas

```python
# Bibliotecas essenciais
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings

# Configurações de visualização
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("Set2")
%matplotlib inline

# Configuração para exibir todas as colunas
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print("✓ Bibliotecas importadas com sucesso!")
```

## 2. Carregamento dos Dados

```python
# URL do dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"

# Nomes das colunas
column_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 
                'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']

# Carregar dados
df = pd.read_csv(url, names=column_names, na_values='?')

# Simplificar target para binário (0 = sem doença, 1 = com doença)
df['target'] = df['target'].apply(lambda x: 0 if x == 0 else 1)

print("✓ Dados carregados com sucesso!")
print(f"Dimensões do dataset: {df.shape[0]} linhas × {df.shape[1]} colunas")
```

## 11. Análise de Outliers

```python
# Boxplots das variáveis numéricas
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.ravel()

for idx, (col, titulo) in enumerate(zip(numericas, titulos_num)):
    data_clean = df[col].dropna()
    axes[idx].boxplot(data_clean, vert=True, patch_artist=True,
                     boxprops=dict(facecolor='lightcoral', color='darkred'),
                     medianprops=dict(color='blue', linewidth=2),
                     whiskerprops=dict(color='darkred'),
                     capprops=dict(color='darkred'))
    axes[idx].set_title(f'Boxplot de {titulo}', fontsize=11, fontweight='bold')
    axes[idx].set_ylabel(titulo, fontsize=10)
    axes[idx].grid(axis='y', alpha=0.3)

# Remover subplot extra
fig.delaxes(axes[5])

plt.tight_layout()
plt.show()
```

```python
# Contagem de outliers (método IQR)
print("=== DETECÇÃO DE OUTLIERS (Método IQR) ===")
print("\nNúmero de outliers por variável:")

for col in numericas:
    data_clean = df[col].dropna()
    Q1 = data_clean.quantile(0.25)
    Q3 = data_clean.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data_clean[(data_clean < lower_bound) | (data_clean > upper_bound)]
    
    print(f"{col:15s}: {len(outliers):3d} outliers ({len(outliers)/len(data_clean)*100:.1f}%)")
```
